```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

####**Data Reading**

```python
df=spark.read.format('parquet').load('abfss://bronze@databrickseteprojectsa.dfs.core.windows.net/customers')
df.display()
```

#####Drop rescued data

```python
df=df.drop('_rescued_data')
```

#####Create domains columns from email

```python
df=df.withColumn('domains',split(col('email'),'@')[1])
df.display()
```

#####create fullname column

```python
df=df.withColumn('Full_name',concat('first_name',lit(' '),'last_name'))
df=df.drop('first_name','last_name')
df.display()
```

####Data writing

```python
df.write.format('delta').mode('append').save('abfss://silver@databrickseteprojectsa.dfs.core.windows.net/customers')
```
